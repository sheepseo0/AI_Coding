```python
a = int(input("초를 입력하세요 :"))
sec = a
hour = sec//3600
minute = sec%3600//60
second = sec%3600%60

print(f"{hour}시간{minute}분{second}초")  #f스트링 쓰는 법 알기
```

```python
americano_price = 2000
cafelatte_price = 3000
capucino_price = 3500

americanos = int(input("아메리카노 판매 개수: "))
cafelattes = int(input("카페라떼 판매 개수: "))
capucinos = int(input("카푸치노 판매 개수: "))

sales = americanos*americano_price
sales = sales + cafelattes*cafelatte_price
sales = sales + capucinos*capucino_price

print("총 매출은",sales,"입니다.")
```

```python
# LAB 3-3. 자동 판매기 프로그램 만들기

money = int(input("투입한 돈: ")) #5000원을 받고,
price = int(input("물건 값: ")) #물건 값은 2600원이고, 
change = money-price #거스름돈을 500, 100으로 나눠야하는데..........??

print(f"거스름돈:{change}")
coin500 = change//500
change = change%500
coin100 = change//100

print(f"500원 동전의 개수:{coin500}")
print(f"100원 동전의 개수:{coin100}")
```
