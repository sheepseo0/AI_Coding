```python
num1 = int(input("첫 번째 정수를 입력하시오 :"))
num2 = int(input("두 번째 정수를 입력하시오 :"))
num3= int(input("세 번째 정수를 입력하시오 :"))
avg = (num1 + num2 + num3)/3

print(num1,",", num2, ",", num3, "의 평균은", avg, "입니다.")
```

```python
x = int(input("첫 번째 정수를 입력하시오: "))
y = int(input("두 번째 정수를 입력하시오: ")) 

result = x + y
print(x, "+", y, "=", result)

result = x - y
print(x, "-", y, "=", result)

result = x * y
print(x, "*", y, "=", result)

result = x / y
print(x, "/", y, "=", result)
```
