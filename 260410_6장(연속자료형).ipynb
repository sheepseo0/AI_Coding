```python
#1~100까지의 짝수 리스트 만들기
even = []
for i in range(1,101):
    if i % 2 == 0:
        even.append(i) #even 리스트의 마지막 위치에 i 추가

    #print("1~100 까지의 짝수 리스트:", even) #이걸 for문 안에 넣으면 리스트가 추가되는 과정이 다 출력됨
#print("1~100 까지의 짝수 리스트:", even) #밖으로 빼면 마지막 최종 결과값만 보임

print(even)
```

```python
#딕셔너리
food = {'춘천':'닭갈비', '횡성':'한우', '부산':'밀면', '전주':'비빔밥'}
print(food["춘천"])

food['춘천'] = '냉면'
print(food)

del food['횡성']
print(food)

del food
print(food)


```

```python
#딕셔너리 전용 함수 연습
food = {'춘천':'닭갈비', '횡성':'한우', '부산':'밀면', '전주':'비빔밥'}
food.get('서울', '내용이 없습니다.')
print(food)

food.pop('횡성')
print(food)

food.keys()

food.values()

food.items()

food.update({'포항':'과매기'})
print(food)

food.clear()
print(food)
```

```python
menu = {'김밥':'3000', '라면':'5000','피자':'10,000'}

order = input("어떤 메뉴를 원하세요?")
if order in menu: #order의 내용이 menu의 key 중 하나일 때만 True
    print(f"가격은 {menu[order]}원입니다.") #지금 여기서 key 대변하는 게 order!
```

```python
#딕셔너리에 반복문
bread = {'왕밤빵':'2000','단팥빵':'2000','에그타르트':'3500'}
for i in bread.keys(): #key의 리스트를 만드는 것
    print(f"{i}의 가격은 {bread[i]}원입니다.") #오잉 어떻게 이렇게 나오지
```

```python
students = [{"학번":202155523,"이름":"김여자","점수":90},
            {"학번":202155578,"이름":"이남자","점수":85}]

for s in students:  # s는 각 학생에 대한 딕셔너리 자료
    print(f"{s['학번']} {s['이름']} 학생은 {s['점수']}점입니다.")
```

```python
#이건 내가 한 거
score = []
for i in range(7):
    a = int(input("1~10점 사이의 점수 입력 :"))
    score.append(a)

score.sort()
s_score=score[1:6]

score_sum = sum(s_score)
avg = sum(s_score)/len(s_score)

print(f"점수 합계={score_sum}, 평균 점수={avg}")
```

```python
#교수님 방식
score = []
for i in range(7):
    score.append(int(input(f"{i+1}번째 심사위원의 점수는?")))
print(f"원점수: {score}")
score.sort()
print(f"정렬된 점수: {score}")
sum = 0
for i in score[1:-1]:
    sum = sum+1

print(f"합계:{sum}, 평균점수:{sum/5}")

#엥 이거 맞음???
```

```python
country_capital = {
    "아이슬란드": "레이캬비크",
    "터키": "앙카라",
    "캐나다": "오타와",
    "뉴질랜드": "웰링턴",
    "벨기에": "브뤼셀",
    "우간다": "캄팔라",
    "파라과이": "아순시온"
}
while True:
    country = input(f"다음 {list(country_capital.keys())} 중 어느 나라의 수도를 알고 싶은가요?") #리스트 함수.. 

    if country in country_capital:
        print(f"{country}의 수도는 {country_capital[country]}입니다.")
    elif country == '끝':
        print(f"프로그램을 종료합니다.")
        break
    else:
        print("목록에 없는 나라입니다.")
```
