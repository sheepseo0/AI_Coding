```python
def addition(num1, num2):
    return num1+num2


def sutraction(num1, num2):
    if num1>num2:
        return num1-num2
    else:
        return num2-num1


def multiplication(num1, num2):
    return num1*num2

def devision(num1, num2):
    if num2 ==0:
        return 0
    else:
        return num1/num2

def power(num1, num2):
    result = 1
    for i in range(num2):
        result = result*multiplication(result, num1)
    return result

power(2,3)
```
