```python
떡볶이 = 2000
라면 = 3000
튀김 = 500
떡볶이_판매 = int(input("떡볶이를 몇개나 팔았습니까?"))
라면_판매 = int(input("라면을 몇개나 팔았습니까?"))
튀김_판매 = int(input("튀김를 몇개나 팔았습니까?"))
떡볶이매출 = 떡볶이*떡볶이_판매
라면매출 = 라면*라면_판매
튀김매출 = 튀김*튀김_판매

총매출 = 떡볶이매출+라면매출+튀김매출
print("오늘의 총 매출 :",총매출)
```

```python
p1 = 1300
p2 = 1500
money = int(input("얼마 투입하셨습니까 :"))
buy1 = int(input("p1 구간 티켓을 몇 장 구매하셨습니까? :"))
buy2 = int(input("p2 구간 티켓을 몇 장 구매하셨습니까? :"))
change = money-((p1*buy1) + (p2*buy2))
print("거스름돈 :",change)

change1000 = change // 1000
change = change % 1000
change100 = change // 100

print(f"1000원 :{change1000}장")
print(f"100원 :{change100}개")
```

```python
num = int(input("숫자 입력:"))

if num > 0:
    print("양수입니다.")
    if num%2 == 0:
        print("짝수입니다.")
    else: #num % 2 == 1 
        print("홀수입니다.")
else: #num <= 0
    if num == 0:
        print("0입니다.")
    else: # num < 0
        print("음수입니다.")
```

```python
import turtle

t = turtle.Turtle()
t.shape("turtle")

t.penup() #펜 들어서 그림 그려지지 않음 

t.goto(100, 100) #거북이를 100,100으로 이동시킴
t.write("거북이가 여기로 오면 양수입니다.")
t.goto(100,0)
t.write("거북이가 여기로 오면 0입니다.")
t.goto(100,-100)
t.write("거북이가 여기로 오면 음수입니다.")

t.goto(0,0)
t.pendown()

s = int(turtle.textinput(" ","숫자를 입력하시오:"))
n = int(s)

if n > 0:
    t.goto(100,100)
elif n == 0:
    t.goto(100,0)
else: #n이 0보다 작으면
    t.goto(100,-100)

'''
오류 떴다는데?
복사 하고 다시 해보기
'''
```
