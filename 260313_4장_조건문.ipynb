```python
p1 = 1300
p2 = 1500
money = int(input("얼마 투입하셨습니까 :"))
buy1 = int(input("p1 구간 티켓을 몇 장 구매하셨습니까? :"))
buy2 = int(input("p2 구간 티켓을 몇 장 구매하셨습니까? :"))
change = money-((p1*buy1) + (p2*buy2))
print("거스름돈 :",change)

change1000 = change // 1000
change = change % 1000
change100 = change // 100

print(f"1000원 :{change1000}장")
print(f"100원 :{change100}개")
```

```python
#중첩 if
num = int(input("숫자 입력:"))

if num > 0:
    print("양수입니다.")
    if num%2 == 0:
        print("짝수입니다.")
    else: #num % 2 == 1 
        print("홀수입니다.")
else: #num <= 0
    if num == 0:
        print("0입니다.")
    else: # num < 0
        print("음수입니다.")
```

```python
import turtle

t = turtle.Turtle()
t.shape("turtle")

t.penup() #펜 들어서 그림 그려지지 않음 

t.goto(100, 100) #거북이를 100,100으로 이동시킴
t.write("거북이가 여기로 오면 양수입니다.")
t.goto(100,0)
t.write("거북이가 여기로 오면 0입니다.")
t.goto(100,-100)
t.write("거북이가 여기로 오면 음수입니다.")

t.goto(0,0)
t.pendown()

s = int(turtle.textinput(" ","숫자를 입력하시오:"))
n = int(s)

if n > 0:
    t.goto(100,100)
elif n == 0:
    t.goto(100,0)
else: #n이 0보다 작으면
    t.goto(100,-100)

'''
오류 떴다는데?
복사 하고 다시 해보기
'''
```

```python
score = int(input("점수를 입력하시오 :"))

if score >= 90:
    print("A",end='') #end='' : print에는 end옵션이 들어가있다. 행이 바뀐다는 의미.end뒤에 '' 붙으면 새로운 행으로 바뀌지 않고 바로 옆으로 간다.
else:
    if score >= 80:
        print("B",end='')
    else:
        if score >= 70:
            print("C",end='')
        else:
            if score >= 60:
                print("D",end='')
            else:
                print("F", end='')

print("학점입니다.")
```

```python
score = int(input("점수를 입력 ==> "))
if score >= 90 :
    print("A", end='')
elif score >= 80:
    print("B", end='')
elif score >= 70:
    print("C", end='')
elif score >= 60:
    print("D", end='')
else:
    print("F", end='')
    
print("학점입니다.")
```

```python
#도전문제 4.10

x = int(input("점의 좌표 x를 입력하시오.==>"))
y = int(input("점의 좌표 y를 입력하시오.==>"))

d = ((x-3)**2 + (y-4)**2)**(0.5)
 
if d < 10:
    print("원 내부에 있음")
elif d == 10:
    print("원 둘레에 걸쳐 있음")
else: # d > 10
    print("원 외부에 있음")
```
