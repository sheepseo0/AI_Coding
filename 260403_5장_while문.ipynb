```python
import random
tries = 0
guess = 0
answer = random.randint(1,100)
while True:
    guess = int(input("1부터 100 사이의 숫자를 맞추시오"))
    tries += 1
    if guess > answer:
        print(f"높음!")
    elif guess < answer:
        print(f"낮음!")
    else:
        break

print(f"{answer}, 축하합니다. 시도횟수:{tries}")
```

```python
#이건 교수님

i = 1
sum = 0
while i <= 100:
    if i %2==0:
        print(f"{i}", end=',')
        sum = sum +i #if문 안에 있어야 짝수의 합을 구하는 것
    #sum = sum +i => 여기서 하면 1~100까지의 합계가 나옴 
    i = i+1
print(f"\n1~100까지 짝수의 합계:{sum}") #\n : 줄바꿈      
```

```python
while True:
    op = input("사칙연산자 (+, -, *, /) 또는 exit 입력: ")
    if op == 'exit':
        print("프로그램 종료")
        break

    if op not in ['+', '-', '*', '/']:
        print("연산자 오류입니다. 다시 입력하세요.")
        continue

    try:
        num1 = float(input("첫 번째 숫자 입력: "))
        num2 = float(input("두 번째 숫자 입력: "))
    except ValueError:
        print("숫자 입력이 올바르지 않습니다. 다시 시도하세요.")
        continue

    if op == '/' and num2 == 0:
        print("나눗셈에서 0으로 나눌 수 없습니다.")
        continue

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        result = num1 / num2

    print(f"결과: {result}")
```
